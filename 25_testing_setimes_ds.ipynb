```python
df = pd.read_json("setimes_dataset/SETimes.json")
df.head()
```

```python
sr = df.loc[df.language == "sr", :].reset_index(drop=True)
hr = df.loc[df.language == "hr", :].reset_index(drop=True)
bs = df.loc[df.language == "bs", :].reset_index(drop=True)
```
